```python
import pandas as pd
```

```python
df = pd.read_csv('gapminder.csv')
```

```python
df.head()
```

```python
#groupby year and show mean of lifeExp

df.groupby(['year'])['lifeExp'].mean()
```

```python
#another way of doing using diff lib

import numpy as np
df.groupby(['year'])['lifeExp'].agg(np.mean)
```

```python
#the year is the index now

df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].agg(np.mean)
```

```python
#to get the normal indices apply reset_index()

df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].agg(np.mean).reset_index()
```

```python
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
#Filter rows by smoker == 'No' and total_bill >= 10

tips.loc[(tips['smoker'] == 'No') & (tips['total_bill'] >= 10)]
```

```python
#What is the average total_bill for each value of smoker, day, and time

tips.groupby(['smoker', 'day', 'time'])['total_bill'].mean().reset_index()
```
